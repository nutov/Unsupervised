### <img src="https://img.icons8.com/color/48/000000/code.png" style="height:50px;display:inline"> Task 1 - Preprocessing
We don't actially care about the counties, and thus we will perform the following preprocessing steps:
* For each state, remove counties that has NaN in at least one of their fields (use `df.dropna()`)
* For each state, *sum* all the countable fields (votes, total population, cvap...)
* For each state, first change percentage fields (*_pct*) to numbers, and then *sum* for each state and finally re-calculate the percentage out of the new total population. Note that the percentage is out of the *total_population*.
* For each state, calculate the *weighted median* of the the household median and then calculate the mean (average of medians)
    * We want the household median for each state, and in order to that we need to compute the weighted median of each county ($median * \text{total_population}$) and then the best we can do (as we don't that data available) is calculate the mean.
* Add a new column called `trump16_pct`- the precentage of votes Trump got in each state.
* You should end up with a `DataFrame` similar to the original (cell above), with the extra column, but with less rows. The name of the new DataFrame should be `prep_df` instead of `df`.

**Hints**: `pandas` has great tools for doing all of this. Check out: `df[df['state']=='Alabama']`, `df.uniuqe()`, `df.sum()`, `df.append()`...

```python
# your code here - you can use as many cells as you need
```

```python
df.dropna(inplace = True)
# Group by state and sum all the countable fields
countable_fields = ['state','total_population', 'cvap','trump16','clinton16','otherpres16']
group_by_state_countable_fields = df[countable_fields].groupby('state')
group_by_state_countable_fields = group_by_state_countable_fields.sum()

# Change percentage fields to numbers 
# and group by state and sum 
df_pct_to_num = df.copy()
pct_index = ['state','white_pct', 'black_pct', 'hispanic_pct',
       'nonwhite_pct', 'foreignborn_pct', 'female_pct',
       'clf_unemploy_pct', 'lesshs_pct', 'lesscollege_pct']
for idx in pct_index:
    if idx == 'state':
        continue
    df_pct_to_num[idx] = (df_pct_to_num[idx]/100)*df_pct_to_num['total_population']
    df_pct_to_num[idx] = df_pct_to_num[idx]//1
    
group_by_state_pct_fields = df_pct_to_num[pct_index].groupby('state')
group_by_state_pct_fields = group_by_state_pct_fields.sum()

# re calculating new pecentage out of the new total total population 
total_population_per_state_df = group_by_state_countable_fields['total_population']

for idx in pct_index:
    if idx == 'state':
        continue
    group_by_state_pct_fields[idx] = 100*group_by_state_pct_fields[idx]/total_population_per_state_df
```

```python
# Merging the new dataframes 
# group_by_state_pct_fields, group_by_state_countable_fields

left = group_by_state_countable_fields
right = group_by_state_pct_fields

```

```python
# Calculate wheighted median
df_median = df[['state','median_hh_inc']].copy()
# median_hh_inc * total_population 
df_median['median_hh_inc'] = df_median['median_hh_inc']*df['total_population']
# devide by the states total population
state_list = list(df['state'].unique())
df_median = df_median.groupby('state').sum()

```

```python
# calculate the weighted median
df_median['median_hh_inc'] = df_median['median_hh_inc']/group_by_state_countable_fields['total_population']
```

```python
left['median_hh_inc'] = df_median['median_hh_inc']
```

```python
# join left and right dataframes to create prep_df
prep_df = left.join(right).copy()
prep_df['trump16_pct'] = 100*prep_df['trump16']/prep_df['total_population']

```

### <img src="https://img.icons8.com/color/48/000000/code.png" style="height:50px;display:inline"> Task 2 - Point Estimates
As learned in class, point estimates are estimates of population parameters based on sample data. For instance, if we wanted to know the average age of registered voters in the U.S., we could take a survey of registered voters and then use the average age of the respondents as a point estimate of the average age of the population as a whole. The average of a sample is known as the sample mean. The sample mean is usually not exactly the same as the population mean. This difference can be caused by many factors including poor survey design, biased sampling methods and the randomness inherent to drawing a sample from a population

In this task, we will go back to the original data, `df` in the code, and we will treat each *county* as one vote and we will try to estimate the **mean** *median income* of voters.

* Constant the seed (`np.random.seed(0)`)
* Clean the original data by removing counties with NaN (as before, use `df.dropna()`)
* Calculate the **true** *median income* (print it)
* Take a random sample of $n=1000$ samples, calculate the sample mean, and calculate the difference from the true mean (print them)
* Repeat the process of sampling $n=1000$ (with replacement) $N=10, 20, 40, 50, 100, 200, 500, 1000, 2000$ times, and for each $N$ calculate the mean, calculate the difference from the true mean, plot the difference from the true mean vs. $N$.

```python
# your code here - you can use as many cells as you need
np.random.seed(0)
```

```python
df.dropna(inplace = True)

medians = df['median_hh_inc'].values
true_mean = np.mean(medians)
print('The true median income is : {}'.format(true_mean))
```

```python
n = 1000
samp_med_list  = medians[np.random.choice(n,len(medians))]
samp_mean = np.mean(samp_med_list)
true_mean = np.mean(medians)
print('the diff between the true mean and sample mean : {}'.format(true_mean - samp_mean))
```

```python
N_list = [10,20,40,50,100,200,500,1000,2000]
mean_list = []
std = []
for N in N_list:
    T = []
    for _ in range(N):
        samp_med_list  = medians[np.random.choice(len(medians),n)]
        T.append(np.mean(samp_med_list))
    
    mean_list.append(np.mean(T))
    std.append(np.std(T))
    
diff = true_mean - np.array(mean_list) 
```

```python
fig,axes = plt.subplots(nrows = 1,ncols=1,figsize=(7,7))
axes.plot(N_list,np.abs(diff),ls = '-',marker='*')
axes.set_title('Absulute value of the diff between sample mean and true mean')
axes.set_xlabel('N')
axes.set_ylabel('|diff|')
```

### <img src="https://img.icons8.com/color/48/000000/code.png" style="height:50px;display:inline"> Task 3 - Confidence Intervals
A point estimate can give you a rough idea of a population parameter like the mean, but estimates are prone to error and taking multiple samples to get improved estimates may not be feasible (would you call 500 people, 2000 times?). A confidence interval is a range of values above and below a point estimate that captures the true population parameter at some predetermined confidence level. For example, if you want to have a 95% chance of capturing the true population parameter with a point estimate and a corresponding confidence interval, you'd set your confidence level to 95%. Higher confidence levels result in a wider confidence intervals.

Calculating a confidence interval -  taking a point estimate and then adding and subtracting a margin of error to create a range. Margin of error is based on your desired confidence level, the spread of the data and the size of your sample. The way you calculate the margin of error depends on whether you know the standard deviation of the population or not.

* Calculate a 90%, 95% and a 99% confidence interval for the **mean** *median income* of voters. Take a sample of $n=1000$.
    * Equivalently, $\alpha=[0.1, 0.05, 0.01]$
    * Don't forget to **standartize** the data. and notice that $\sigma \neq 1$ as in the tutorial.
    * Use the true STD of the population
    * To calculate the inverse of the CDF, use `scipy.stats.norm.ppf()`
* Compare your answer to using `scipy.stats.norm.interval(alpha=, loc=, scale=)`. Read the doc, use Shift + Tab inside the parenthesis to understand how to use it.
    * You need to print the result for each $\alpha$, with both methods

```python
# your code here - you can use as many cells as you need
np.random.seed(0)
```

```python
# We used the CLT theory
n = 1000
samp_med_list  = medians[np.random.choice(n,len(medians))]
true_std = np.std(medians)
true_mean = np.mean(medians)
est_mean = np.mean(samp_med_list)

aplha_list = [0.1, 0.05, 0.01]
z_alpha_list = []
CI_list = []
for alpha in aplha_list:
    z_alpha_list.append((scipy.stats.norm.ppf(alpha/2),scipy.stats.norm.ppf(1 - alpha/2)))
    CI_list.append(scipy.stats.norm.interval(alpha=alpha, loc=np.mean(samp_med_list), scale=true_std/np.sqrt(n)))
    lower = est_mean +  scipy.stats.norm.ppf(alpha/2)*true_std/(np.sqrt(n))
    upper = est_mean +  scipy.stats.norm.ppf(1-alpha/2)*true_std/(np.sqrt(n))
    print("CI using scipy ppf is, alpha = {alpha} : ({lower} , {upper})".format(lower = int(lower*100)/100.0,upper = int(upper*100)/100.0,alpha = alpha))
    print("Size of CI using scipy ppf , alpha = {alpha} , size = {size}".format(alpha = alpha, size = int(100*(upper-lower)/100.0 )))
    print("\n")
    lower,upper = scipy.stats.norm.interval(alpha=1-alpha, loc=np.mean(samp_med_list), scale=true_std/np.sqrt(n))
    print("CI using scipy interval is, alpha = {alpha} : ({lower} , {upper})".format(lower = int(100*lower)/100.0,upper = int(100*upper)/100.0,alpha = alpha))
    print("Size of CI using scipy interval , alpha = {alpha} , size = {size}".format(alpha = alpha, size = int(100*(upper-lower)/100.0 )))
    print("\n")
    
```

### <img src="https://img.icons8.com/color/48/000000/code.png" style="height:50px;display:inline"> Task 5 - Hypothesis Testing - Two-Sided One Sample T-Test
Point estimates and confidence intervals are basic inference tools that form the foundation for another inference technique: statistical hypothesis testing. Statistical hypothesis testing is a framework for determining whether observed data deviates from what is expected. The T-test is a statistical test used to determine whether a numeric data sample differs significantly from the population or whether two samples differ from one another.

In this exercise we will perform the t-test for the mean of unemployment fraction (`clf_unemploy_pct`) in the US. We will use the pre-procesesd data (`prep_df`) and a sample of size $n=10$. We wish to test whether the sample mean differs from the population mean. The null hypothesis, $H_0$ is that the mean is not different. Recall that we are performng t-tests when we are not given the true mean and true STD. The steps:

* Constant the seed
* Calculate the true mean of unemployment fraction (use fractions and not percents: `clf_unemploy_pct / 100`)
    * In t-test, you assume that the mean ($\mu$) is unknown, then you formulate the $H_0: \mu= \mu_0$. So if $\mu$ is unknown, how do you calculate the t-statistic? Well, assume that somebody told you that in the last elections, $\mu = C$, where $C$ is the also the true mean of this election.
    * In other words, to calculate t-statistic, use the true mean. 
* Take a sample of $n=10$ and calculate its mean.
* Calculate the sample's unbiased STD.
* Calculate the t-statistic value.
    * Use the true mean for $\overline{\mu} - \mu_0$
* Calculate $t_{n-1, \frac{\alpha}{2}}$ for $\alpha=[0.1, 0.05, 0.01]$ significance levels (how many derees of freedom are there for the t-distribution?). Use `scipy.stats.t.ppf()`.
* Calculate the p-value (don't forget that this is a **two-sided** test). For each level, do we reject $H_0$ or accept it?. Use `scipy.stats.t.cdf()`.
* Compare your results to `scipy.stats.ttest_1samp()`. This function performs the t-test, and outputs the t-statistic and the p-value.

Note: "Calculate" = print the results

* Tips:
    * When calculating the p-value, pay attention to which side are you on (if t-statitic > 0, then you should calculate the p-value for $1-\phi(x)$)
    * Your results should align with the output of `scipy.stats.ttest_1samp()` (make sure the seed is constant)

```python
# your code here - you can use as many cells as you need
np.random.seed(0)
```

```python
def GetThreshold(alpha:float,n:int):
    return scipy.stats.t.ppf(1-0.5*alpha,n-1)
```

```python
# Get the data as np array
n=10
unemployment_np = df['clf_unemploy_pct'].values
unemployment_np = unemployment_np/100
# Caculate the true mean of the data
true_mean_unemployment = np.mean(unemployment_np)
# Sampling
unemployment_np_samp = unemployment_np[np.random.choice(len(unemployment_np),n)]
# Calculating the sample mean and the unbiased std
samp_mean = np.mean(unemployment_np_samp)
samp_ub_std = np.sqrt(n/(n-1))*np.std(unemployment_np_samp)

t_test = samp_mean - true_mean_unemployment/samp_ub_std
alpha_list = [0.1,0.05,0.01]

p_val = 2*scipy.stats.t.cdf(t_test,df=n-1) if t_test < 0 else 2*(1-scipy.stats.t.cdf(t_test,df=n-1))

for alpha in alpha_list:
    print('the treshold is : {}'.format(GetThreshold(alpha,n)))
    if np.abs(t_test) <= GetThreshold(alpha,n):
        print('H0 is accepted for alpha = {}'.format(alpha))
    else:
        print('H0 is rejected for alpha = {}'.format(alpha))
    
    
print('p-value is : {}'.format(p_val))
```

### <img src="https://img.icons8.com/cute-clipart/64/000000/wine-glass.png" style="height:50px;display:inline"> Part 2 - Dimensionality Reduction - Prologue - Wine Dataset
In this exercise we are going to compare different dimensionality reduction techniques on the Wine dataset.
The wine dataset is a classic and very easy multi-class classification dataset. There are 3 types of wine, 178 examples with 13 features each. Even though it is a very dataset for classification, we will do unsupervised analysis to compare different aspects of dimensionality reduction methods. Let's look at the data.

```python
# imports for the exrcise - part 2
# you can add more if you wish (but it is not really needed)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA, KernelPCA
from sklearn.manifold import TSNE, LocallyLinearEmbedding, Isomap
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_wine, load_digits
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, Perceptron
```

```python
from sklearn.datasets import load_wine
X, y = load_wine(return_X_y=True)
pd.DataFrame(np.concatenate((X, y.reshape(-1, 1)), axis=1), columns=list(range(13)) + ['class'])
```

### <img src="https://img.icons8.com/color/48/000000/code.png" style="height:50px;display:inline"> Task 6 - Importance of Feature Scaling
* Perform PCA on the data (`X`) to `n_components=2` and plot it.
* Scale the data using `StandardSaler()` to create `X_scaled`, perform PCA on `X_scaled` to `n_components=2` and plot it.

Note: for all algorithms, use constant seed (`random_state=...`), for example: `PCA(n_component=2, random_state=random_state)`

```python
# your code here - you can use as many cells as you need
random_state = 42
np.random.seed(random_state)
```

```python
# define pca class and perform dim reduction 
# on scaled and not scaled datasets 
pca_solver = PCA(n_components=2,random_state=random_state)
scale_solver = StandardScaler()
X_scaled = scale_solver.fit_transform(X)
x_scaled_dim2 = pca_solver.fit_transform(X_scaled)
x_dim2 = pca_solver.fit_transform(X)

#plotting the scaled and unscaled pca

fig,axes = plt.subplots(nrows = 1,ncols=2,figsize=(10,5))
axes[0].scatter(x_scaled_dim2[:,0],x_scaled_dim2[:,1],c=y)
axes[0].set_title('dim reduction using pca after scaling')

axes[1].scatter(x_dim2[:,0],x_dim2[:,1],c=y)
axes[1].set_title('dim reduction using pca without scaling')
plt.tight_layout()
```

### <img src="https://img.icons8.com/bubbles/48/000000/ask-question.png" style="height:50px;display:inline"> Question 6 - Importance of Feature Scaling
- Why is it so important to perform feature scaling before performing dimensionality reduction, espicially for PCA?
- Describe the results of Task 6

## answer:
In pca algorithm we want to find the k principle components that have the largest variance,we can view pca as finding the  centered ellipsoid that best surrounds the data, and the k principle components are the k largest radiuses of the ellipsoid. 

If the data is not centered then the ellipsoid surrounds biased datapoints and a large radius of the ellipsoid could be due to a big distance from the center and not due to a big variance of the data.

Dividing by the variance is not mandatory but depends on the dataset.

- Explain in short how t-SNE works and how it is different from PCA.
- Compare the results of PCA nad t-SNE on the Wine dataset

#### Answer 7

1. The t-sne algorithm models a probality sidtribution of neighbors around each point.<br>
In the high dimentional space the distance is modeled as a guassian distribution, and in the low dimentionmal space the distance is modeled as a t distribution.<br>
The algorithm tries to find a mapping between the high dimentional space (X) and the low dimentional space (y)<br>
that minimizesthe differences between these two distributions over all points. The fatter tails of a t-distribution compared to a Gaussian help to spread the points more evenly in the low dimentional space.<br>

2. We can see that with t-sne algorithm the clusters are more distinguishable.<br>
In both algorithms we can see separated clusters of the results, but if we didnt have labels we could distinguish between<br> the different labels only with the t-sne results because the different clusters are more far apart.

### <img src="https://img.icons8.com/bubbles/48/000000/ask-question.png" style="height:50px;display:inline"> Question 10 - Comparing Dimensionality Reduction Methods
* Shortly describe what each method from the above table does, and the effect of the hyper-parmeters.
* Which method produced the most satisfying results?

#### Answer 10

PCA: Performs a linear projection on the directions which the variance of the data is maximal.<br>

KPCA: First we perform a non-linear mapping to a higher dimention on each data point, the mapping $\phi(x):R^D \to R^M$ s.t $M>>D$. We denote $K_{i,j} = \phi^{T}(x_i) \cdot \phi(x_j)$ , such that $K \in R^{nxn}$, 
after that we centrelize $K$ and we find the directions which the variance of the projected data is maximal.
The matrix $K$ is the kernel matrix , to avoid a computational disaster we do not compute $\phi(x)$ we compute only $K$.<br>

Isomap: We assume that our data sits on a lower dimention manifold than the numbers of features of the data,
so in order to preserve the structure of the data when we perform DR we want to use the MDS algorithm and that the distance between two projected points will be close to the geodisic distance between to data points on the manifold.
We approximate the geodisic distance between two data points on the manifold by building a $k$  nearest neighbors graph
such that each point is connected to its $k$ nearest neighbors by euclidean distance.
We estimate the geodisic distance by finding the shortest path between two points on the graph.<br>
The parameter $k$ controls how much neighbors each points have on the graph if $k$ is to small the graph might be disconnected, if $k$ is to large then our estimation of the geodisic distance will not be accurate.<br>

LLE: Tries to describe the structure of each data point by a linear combination of its $k$ nearest neighbors, 
we get the projection by constructing the points in the projection dimention with the same combination. <br>

T-SNE: A DR technique that tries to perserve the local structure of the data.
The algorithm converts distances to probabilities and tries to perserve the same distribution in the projection dimention.

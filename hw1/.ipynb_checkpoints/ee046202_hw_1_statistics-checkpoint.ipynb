## <img src="https://img.icons8.com/bubbles/48/000000/ask-question.png" style="height:50px;display:inline"> Question 1 - Gaussian RV- Basics
1. Let $Z \sim \mathcal{N}(0,1)$ be a normal Gaussian RV, and $X \sim \mathcal{N}(\mu, \sigma)$. The Cumulative Distribution Function (CDF) of $Z$ is defined as $$ P(Z\leq c) \triangleq \phi(c).$$ Express $P(X\geq x)$ using $\phi(c)$.
2. Consider a sequence of $N$ i.i.d. RVs $\{X_i\}_{i=1}^N$, where $X_i \sim \mathcal(10,1)$. The empirical mean is given by $\overline{X}_N = \frac{1}{N}\sum_{i=1}^NX_i$. What is the distribution of $\overline{X}_N$?
3. What is the probability $P(9.7 \leq \overline{X}_N\leq 10.3)$ for $N=1,10,20$? Express first using the function $\phi(X)$ and then use `scipy.stats.norm.cdf` to calculate $\phi(x)$ and obtain a numerical value.
4. Since Gaussian RVs are not bounded, we cannot use **Hoeffding's** inequality to bound terms of the form  $P(9.7 \leq \overline{X}_N\leq 10.3)$. A possible alternative for this is to use the following proposition (which is more general and holds for a sum of sub-Gaussian RVs):
    * **Proposition**: Let $\{X_i\}_{i=1}^N$ be i.i.d. RVs with $X_i \sim \mathcal{N}(\mu, \sigma)$. Then: $$ P(|\frac{1}{N}\sum_{i=1}^NX_i -\mu|\geq \epsilon) \leq 2 \exp(-\frac{N\epsilon^2}{2\sigma^2}).$$ Use this proposition to find a lower bound for $P(9.7 \leq \overline{X}_N\leq 10.3)$ for $N=1,10,20$. 

#### Answer 1


1. We will denote that $\frac{X-\mu}{\sigma} \sim \mathcal{N}(0,1)$ hence $ P(X \leq x) = P(X - \mu \leq x - \mu) 
= P(\frac{X - \mu}{\sigma} \leq \frac{x - \mu}{\sigma}) = \phi(\frac{x - \mu}{\sigma})$.
so we can infer that $P(X \ge x) = 1 - P(X \le x) = 1 - \phi(\frac{x - \mu}{\sigma}) $
2. The gaussian RVs $\{X_i\}_{i=1}^N$ are i.i.d hence their sum is a gaussian RV $\Rightarrow$ $\overline{X}_N$ is a gaussian RV.

Lets compute $\overline{X}_N$ mean and std:
$ \mu_{new} = E(\overline{X}_N) = E(\frac{1}{N}\sum_{i=1}^{N}X_i) = \frac{1}{N}\sum_{i=1}^{N}E(X_i) = \frac{1}{N}N10 = 10 $,

$ var(\overline{X}_N) = \frac{1}{N^2}var(\sum_{i=1}^{N}X_i) $ from the i.i.d property $\frac{1}{N^2}var(\sum_{i=1}^{N}X_i) = \frac{1}{N^2}\sum_{i=1}^{N}var(X_i) = \frac{1}{N^2}Nvar(X_1) = \frac{1}{N} $,

hence $ \sigma_{new} = \frac{1}{\sqrt{N}} $ $\Rightarrow $ $ \overline{X}_N \sim \mathcal{N}(10,\frac{1}{\sqrt{N}} )$

3. $P(9.7 \le \overline{X}_N \le 10.3) = P(\overline{X}_N \le 10.3) - P(\overline{X}_N \le 9.7) = \phi(0.3\sqrt{N}) - \phi(-0.3\sqrt{N}) $

```python
# imports for q-4
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats  import norm
```

```python
# answer q4-1
N = 100
samples = np.random.randn(N,1)
x = np.linspace(-4, 4, 100)
mu = 0
sigma = 1
y_cdf = norm.cdf(x, mu, sigma) # the normal cdf
hist, bin_edges = np.histogram(samples,bins= np.linspace(-4, 4, N+1))
F_hat = (1/len(samples))*np.cumsum(hist)
alpha = 0.05
epsilon = np.sqrt(np.log((2/alpha))/(2*N))

# plotting the normal CDF and the CDF estimation 
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(1, 1, 1)
ax.plot(x, y_cdf, label='CDF')
ax.plot(np.linspace(-4,4,N),F_hat,label = 'F_hat')
ax.plot(np.linspace(-4,4,N),F_hat-epsilon,label = 'Lower CI')
ax.plot(np.linspace(-4,4,N),F_hat+epsilon,label = 'Upper CI')

ax.grid()
ax.legend()
ax.set_title("normal CDF and its estimation")

```

```python
# computes and returns the estimation for the normal CDF
def Estimate(N = 100,alpha = 0.05):
    samples = np.random.randn(N,1)
    x = np.linspace(-4, 4, 100)
    mu = 0
    sigma = 1
    y_cdf = norm.cdf(x, mu, sigma) # the normal cdf
    hist, bin_edges = np.histogram(samples,bins= np.linspace(-4, 4, N+1))
    F_hat = (1/len(samples))*np.cumsum(hist)
    epsilon = np.sqrt(np.log((2/alpha))/(2*N))
    return F_hat,epsilon
    
# Returns true if the CDF is in the CI returns false otherwise    
def IsInConfidenceInterval(F_hat,y_cdf,epsilon):
    is_in_interval = np.less_equal(y_cdf, F_hat+epsilon).all() & np.greater_equal(y_cdf, F_hat - epsilon).all()
    return is_in_interval

def ComputeEstimationScore(F_hat,y_cdf):
    diff = np.abs(F_hat-y_cdf)
    return np.amax(diff)
```

```python
K = 1000
count_in_interval = 0

best_F_hat_score = np.Inf
worst_F_hat_score = -np.Inf
mu = 0
sigma = 1
y_cdf = norm.cdf(x, mu, sigma) # the normal cdf

# Finding best and worst estimations
# and percentage of CI
for n in range(0,K):
    F_hat,epsilon = Estimate(N = 100,alpha = 0.05)
    curr_score  = ComputeEstimationScore(F_hat,y_cdf)
    if IsInConfidenceInterval(F_hat,y_cdf,epsilon):
        count_in_interval = count_in_interval+1
   
    if curr_score < best_F_hat_score:
        best_F_hat_score = curr_score
        F_hat_best = F_hat
    
    if curr_score > worst_F_hat_score:
        worst_F_hat_score = curr_score
        F_hat_worst = F_hat

percentage = count_in_interval/K
print(" The percentage wich the estimation was in the CI is : {:.3f} ".format(100*percentage))

fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(1, 1, 1)
ax.plot(x, y_cdf, label='CDF')
ax.plot(np.linspace(-4,4,N),F_hat_best,label = 'best F_hat')
ax.plot(np.linspace(-4,4,N),F_hat_worst,label = 'worst F_hat')


ax.grid()
ax.legend()
ax.set_title("Best and worst estimations")
```
